```python
Q1=df1['Salary'].quantile(0.25)
```

```python
Q3=df1['Salary'].quantile(0.75)
```

```python
IQR=Q3-Q1
```

```python
lb=Q1-0.5*IQR
```

```python
ub=Q3+1.5*IQR
```

```python
outlier=((df1[['Salary']]<lb) | (df1[['Salary']]>ub))
```

```python
clean_ed=df1[~outlier]
```
