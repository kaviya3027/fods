```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
```

```python
data=pd.read_csv("inventory_dataset.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data['Purchase_Price'].fillna(data['Purchase_Price'].median())
```

```python
data['Purchase_Price'].median()
```

```python
data['Sale_Price'].fillna(data['Sale_Price'].mean())
```

```python
data['Sale_Price'].mean()
```

```python
data.dropna()
```

```python
data
```

```python
Q1 = data['Sale_Price'].quantile(0.25)
Q2 = data['Sale_Price'].quantile(0.75)
IQR = Q2 - Q1
```

```python
IQR
```

```python
low = Q1 - 1.5 * IQR
up = Q2 + 1.5 * IQR
```

```python
low
```

```python
up
```

```python
data = data[(data['Sale_Price'] >= low) & (data['Sale_Price'] <= up)]
```

```python
data
```

```python
scaler = MinMaxScaler()
data.loc[:, 'Quantity_In_Stock'] = scaler.fit_transform(data[['Quantity_In_Stock']])
```

```python
data.loc[:, 'Quantity_In_Stock']
```

```python
data.loc[:, 'Quantity_In_Stock'] = (data['Quantity_In_Stock'] * 100).fillna(0).astype(int)
```

```python
data.loc[:, 'Quantity_In_Stock']
```

```python
label_encoder = LabelEncoder()
data.loc[:, 'Discontinued'] = label_encoder.fit_transform(data['Discontinued'])
```
